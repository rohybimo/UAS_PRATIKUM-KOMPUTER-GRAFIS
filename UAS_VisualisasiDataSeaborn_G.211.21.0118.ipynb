```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dt = pd.read_csv('https://raw.githubusercontent.com/rohybimo/uasgrafis/main/dataset/stock.csv')
dt.head(13)
```

```python
# Distribusi df pada setiap atribut Purchase Price, Current Price, dan Gain/Loss dengan grafik histogram
plt.figure(figsize=(10, 4))

plt.subplot(1, 3, 1)
sns.histplot(dt['Purchase Price ($)'], kde=True, bins=10, edgecolor='w', color="k")
plt.xlabel('Purchase Price')
plt.ylabel('Frequency')
plt.title('Histogram of Purchase Price')

plt.subplot(1, 3, 2)
sns.histplot(dt['current Price ($)'], kde=True, bins=10, edgecolor='k', color="y")
plt.xlabel('Current Price')
plt.ylabel('Frequency')
plt.title('Histogram of Current Price')

plt.subplot(1, 3, 3)
sns.histplot(dt['Gain/Loss ($)'], kde=True, bins=10, edgecolor='w', color="r")
plt.xlabel('Gain/Loss')
plt.ylabel('Frequency')
plt.title('Histogram of Gain/Loss')

plt.tight_layout()
plt.show()
```

```python
# Sort data by Dividend / Shares
data = dt.sort_values(by="Dividend/share", ascending=False)
# Get the 5th highest Dividend / Shares
top_5 = data.head(5)
# Plot the Data
plt.figure(figsize=(9, 4))
sns.barplot(x="Stock Name", y="Dividend/share", data=top_5)
plt.title("Top 5 Stocks with Highest Dividend / share")
plt.show()
```

```python
# Sort data by Market Value ($)
data = dt.sort_values(by="Market Value ($)", ascending=False)
# Get the 5th highest Market Value ($)
top_5 = data.head(5)
# Plot the data
plt.figure(figsize=(9, 4))
sns.barplot(x="Stock Name", y="Market Value ($)", data=top_5)
plt.title("Top 5 Stocks with Highest Market Value ($)")
plt.show()
```

```python
# Sort data by Annual Yield (%)
data = dt.sort_values(by="Annual Yield (%)", ascending=False)
# Get the 5th highest Annual Yield (%)
top_7 = data.head(7)
# Plot the data
plt.figure(figsize=(13, 4))
sns.barplot(x="Stock Name", y="Annual Yield (%)", data=top_7)
plt.title("Top 7 Stocks with Highest Annual Yield (%)")
plt.show()
```
